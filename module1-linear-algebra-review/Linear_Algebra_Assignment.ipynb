## 1.1 Create a two-dimensional vector and plot it on a graph

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

vec1 = [[0,0],[1,1]]
plt.arrow(vec1[0][0], vec1[0][1], vec1[1][0], vec1[1][1], head_width=.05, head_length=0.05, color ='red')
plt.xlim(-1,2)          
plt.ylim(-1,2)
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

vectors = np.array([[0, 0, 0, .5, .5, .5]])

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
A = [[1,3,5],[2,4,6]]
A = np.asmatrix(A).T

B = [[2,4,6]]
B = np.asmatrix(B)

C = [[9,6,3],[4,7,11]]
C = np.asmatrix(C)

D = [[1,0,0],[0,1,0],[0,0,1]]
D = np.asmatrix(D)

E = [[1,3],[5,7]]
E = np.asmatrix(E)

print('CD\n', np.multiply(C,D))
print('AE\n', np.multiply(A,E))
print('BA\n', np.multiply(B,A))
```
